```python
#import dependenceis
import pandas as pd
```

```python
#import yearly data from csvs
attendance06 = pd.read_csv("./Data Files/2006.csv")
attendance07 = pd.read_csv("./Data Files/2007.csv")
attendance08 = pd.read_csv("./Data Files/2008.csv")
attendance09 = pd.read_csv("./Data Files/2009.csv")
attendance10 = pd.read_csv("./Data Files/2010.csv")
attendance11 = pd.read_csv("./Data Files/2011.csv")
attendance12 = pd.read_csv("./Data Files/2012.csv")
attendance13 = pd.read_csv("./Data Files/2013.csv")
attendance14 = pd.read_csv("./Data Files/2014.csv")
attendance15 = pd.read_csv("./Data Files/2015.csv")
attendance16 = pd.read_csv("./Data Files/2016.csv")
```

```python
#put data files into list and add a column to each row for the year

years = [attendance06,attendance07,attendance08,attendance09,attendance10,attendance11,attendance12,attendance13,attendance14,attendance15,attendance16]

date = 2006
counter = 0
for year in years:
    items = []
    for x in range(30):
        items.append(date)
    years[counter]["year"] = items
    counter = counter +1
    date = date +1
```

```python
#combine each year of data into a single list

counter = 0
data = years[0]
for things in years:
    data = data.merge(years[counter], how = 'outer')
    counter = counter + 1
```

```python
#merge in last year and then rename columns
data = data.merge(years[1], how = 'outer')

data = data.drop(columns = ['BatAge','PAge','BPF','PPF','#HOF','#A-S','#a-tA-S','Time','Managers','Chall','Succ','Succ%'])

data
```
